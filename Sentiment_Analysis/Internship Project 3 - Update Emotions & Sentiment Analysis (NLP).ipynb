```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd, string

from collections import Counter

import winsound # Play system sound in case of long runtime processes

import nltk
# nltk.download('averaged_perceptron_tagger')
# nltk.download('tagsets')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Text = The Holy Bible - King James Version
text = open('the-holy-bible-king-james-version.txt', encoding='utf-8').read()
lower_text = text.lower()
sentences = sent_tokenize(lower_text)
```

```python
# nltk.help.upenn_tagset() # list all pos
verbs, PoS = [], []
for sentence in sentences:
    for word, pos in nltk.pos_tag(word_tokenize(str(sentence))):
        if (pos=='VB' or pos=='VBD' or pos=='VBG' or pos=='VBN' or pos=='VBP'):
            verbs.append(word)
            PoS.append(pos)
print("Total number of verbs ::", len(verbs))

winsound.Beep(1000,500)
```

```python
set_verbs = set(verbs)
len(set_verbs)
```

```python
lemma_words = []
for word in verbs:
    word = WordNetLemmatizer().lemmatize(word)
    lemma_words.append(word)
```

```python
for i, j in enumerate(lemma_words):
    if j!=verbs[i]:
        print(j,"--->",verbs[i])

winsound.Beep(1000,500)
```

```python
df = pd.DataFrame([verbs, lemma_words, PoS], index=['Words', 'Lemmas', 'PoS']).T
df.head()
```

```python
df1 = df.drop_duplicates(keep='first')
df1.reset_index(inplace=True)
df1.drop(columns=['index'], axis=1, inplace=True)
df1
```

```python
sia = SentimentIntensityAnalyzer()

df1['Sentiment'] = 'Unknown'
for i, j in enumerate(df1['Lemmas']):
    sent = sia.polarity_scores(j)
    if (sent['neg']==1):
        df1['Sentiment'][i]="Negative"
    elif (sent['pos']==1):
        df1['Sentiment'][i]="Positive"
    elif (sent['neu']==1):
        df1['Sentiment'][i]="Neutral"
    else: df1['Sentiment'][i]="Unknown"

winsound.Beep(1000,500)
```

```python
df1['Sentiment'].value_counts()
```

```python
df1 = df1[df1['Sentiment']!='Unknown']
df1.reset_index(inplace=True)
df1.shape
```

```python
df1.drop(columns=['index', 'Words'], axis=1, inplace=True)
df1.head(5)
```

```python
emotion_list = []

with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        if word not in df1['Lemmas']:
            emotion_list.append(word)

w = Counter(emotion_list)
print("Count of each Emotion ::\n", w)
```

```python
# plt.figure(figsize=(16,5))
fig, ax1 = plt.subplots(1,1,figsize=(16,4))
ax1.bar(w.keys(), w.values())
fig.autofmt_xdate()
#plt.savefig('graph.png')
plt.show()
```
