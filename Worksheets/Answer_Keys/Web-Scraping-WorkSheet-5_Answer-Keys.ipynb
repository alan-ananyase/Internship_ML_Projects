```python
#importing required libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
# Activating the web browser
driver = webdriver.Chrome(r'H:/Flip ROBO/chromedriver')
# Opening IMDB website
driver.get('https://www.imdb.com/')
# Locationg Menu
button = driver.find_element_by_id("imdbHeader-navDrawerOpen--desktop")
# Clicking the menu button
button.click()
# Locating Top Rated Movies
top_mov = driver.find_element_by_xpath('//a[@class="ipc-list__item nav-link sc-jTzLTM fjLstn ipc-list__item--indent-one"][2]')
# CLicking Top Rated Movies
top_mov.click()
```

```python
# Scraping movie urls of top 100 movies
mov_links = []                 # List for storing movie urls
for i in range(100):
    try:
        # Scraping urls of i+1 th movie
        link = driver.find_element_by_xpath('//table[@class="chart full-width"]/tbody/tr[{}]/td[2]/a'.format(i+1)).get_attribute('href')
        if link[0:4]=='http':               # Checking if the link is a valid url
            mov_links.append(link)           # appending url to the mov_links list
    except StaleElementReferenceException as e:
        print("Error Occured : ", str(e))   # Handling Exception  
len(mov_links)
```

```python
# Creating dictionary to store movie data
mov_dict = {}
mov_dict["Name"] = []
mov_dict["Year"] = []
mov_dict["IMDB Rating"] = []
mov_dict["Duration"] = []
mov_dict["Director"] = []
mov_dict["Star Cast"] = []
```

```python
for link in mov_links:
    driver.get(link)                                                                          # Loading the webpage by url
    print("Scraping link = ", link)
    time.sleep(2)
    
    try:
        name = driver.find_element_by_xpath('//div[@class="title_wrapper"]/h1')               # Extracting Name from xpath
        mov_dict["Name"].append(name.text.split('(')[0].strip())
    except NoSuchElementException:
        mov_dict["Name"].append('')
    
    try:
        year = driver.find_element_by_xpath('//div[@class="title_wrapper"]/h1/span/a')        # Extracting Year from xpath
        mov_dict["Year"].append(year.text)
    except NoSuchElementException:
        mov_dict["Year"].append('')
        
    try:
        rating = driver.find_element_by_xpath('//div[@class="ratingValue"]/strong/span')      # Extracting rating from xpath
        mov_dict["IMDB Rating"].append(rating.text)
    except NoSuchElementException:
        mov_dict["IMDB Rating"].append('')
        
    try:
        dur = driver.find_element_by_xpath('//div[@class="subtext"]/time')                    # Extracting Duration from xpath
        mov_dict["Duration"].append(dur.text)
    except NoSuchElementException:
        mov_dict["Duration"].append('')
        
    try:
        director = driver.find_element_by_xpath('//div[@class="credit_summary_item"][1]/a')   # Extracting director from xpath
        mov_dict["Director"].append(director.text)
    except NoSuchElementException:
        mov_dict["Director"].append('')
        
    try:
        driver.find_element_by_xpath('//div[@class="credit_summary_item"][3]/a[4]').click()   # Clicking link to open full cast and crew
        time.sleep(2)
        stars = []
        for j in range(10):
            star = driver.find_element_by_xpath('//table[@class="cast_list"]/tbody/tr[{}]/td[2]'.format(j+2))
            stars.append(star.text)
        stars = ', '.join(stars)
        mov_dict["Star Cast"].append(stars)
    except NoSuchElementException:
        mov_dict["Star Cast"].append('')
```

```python
# closing browser
driver.close()
```

```python
movies = pd.DataFrame.from_dict(mov_dict)
movies
```

```python
try:
    url_string = driver.current_url                                   # Scraping url of current page
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)                     # Extracting Latitude and Longitude from url
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")                          # Seperating Latitude and Longitude
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))       # Printing Latitude and Longitude   

except Exception as e:                                                # Handling Exceptions if any
        print("Error: ", str(e))
```

```python
# Closing Web Browser
driver.close()
```

```python
#importing required libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
# Activating the web browser
driver = webdriver.Chrome(r'H:/Flip ROBO/chromedriver')
# Opening Bookpage website
driver.get('https://bookpage.com/')
# Locationg Book Reviews
button = driver.find_element_by_xpath('//li[@class=" dropdown"]/a')
# Clicking the Book Review button
button.click()
```

```python
book_dict = {}
book_dict["Name"] = []
book_dict["Author"] = []
book_dict["Genre"] = []
book_dict["Review"] = []
```

```python
for i in range(5):
    try:
        name = driver.find_element_by_xpath('//div[@class="row-fluid article-row"][{}]/div/div[2]/h4'.format(i+1))
        if name.text[0].isalnum():
            book_dict["Name"].append(name.text)
        else:
            book_dict["Name"].append(name.text.split(' ',1)[1])
    except NoSuchElementException:
        book_dict["Name"].append('-')
    
    try:
        author = driver.find_element_by_xpath('//div[@class="row-fluid article-row"][{}]/div/div[2]/p'.format(i+1))
        book_dict["Author"].append(author.text)
    except NoSuchElementException:
        book_dict["Author"].append('-')
        
    try:
        Genre = driver.find_element_by_xpath('//div[@class="row-fluid article-row"][{}]/div/div[2]/p[2]'.format(i+1))
        book_dict["Genre"].append(Genre.text)
    except NoSuchElementException:
        book_dict["Genre"].append('-')
        
    try:
        current_pg = driver.current_url
        driver.find_element_by_xpath('//div[@class="row-fluid article-row"][{}]/div/div[2]/div/a'.format(i+1)).click()
        time.sleep(3)
        review = driver.find_element_by_xpath('//div[@class="article-body"]')
        book_dict["Review"].append(review.text.replace('\n',' *** '))
        driver.get(current_pg)
        time.sleep(3)
    except NoSuchElementException:
        book_dict["Review"].append('-')
book_dict
```

```python
# Closing the browser
driver.close()
```

```python
books = pd.DataFrame.from_dict(book_dict)
books
```

```python
#importing required libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
# Activating the web browser
driver = webdriver.Chrome(r'H:/Flip ROBO/chromedriver')
# Opening icc-cricket website
driver.get('https://www.icc-cricket.com/')
time.sleep(5)
```

```python
# Locating Team Rankings url from website footer Menu
rank_url = driver.find_element_by_xpath('//ul[@class="global-footer__list"][3]/li[2]/a').get_attribute('href')
# Opening Team Rankings
driver.get(rank_url)
time.sleep(5)
# Locating ODI rankings button
button = driver.find_element_by_xpath('//ul[@class="inline-list"]/li[2]/a')
# Clicking on ODI Button
button.click()
time.sleep(5)
```

```python
# Initialising Team dictionary
teams = {}
teams["Team"] = []
teams["Ranking"] = []
```

```python
# Extracting Top 10 ranked teams
for i in range(10):
    try:
        rank = driver.find_element_by_xpath('//div[@class="rankings-block__container full rankings-table"]/table/tbody/tr[{}]/td[1]'.format(i+1))
        teams["Ranking"].append(rank.text)
    except NoSuchElementException:
        teams["Ranking"].append('-')
    
    try:
        team = driver.find_element_by_xpath('//div[@class="rankings-block__container full rankings-table"]/table/tbody/tr[{}]/td[2]/span[2]'.format(i+1))
        teams["Team"].append(team.text)
    except NoSuchElementException:
        teams["Team"].append('-')
```

```python
# Locating player Rankings url from website footer Menu
rank_url = driver.find_element_by_xpath('//ul[@class="global-footer__list"][3]/li[3]/a').get_attribute('href')
# Opening Player Rankings
driver.get(rank_url)
time.sleep(5)
# Locating ODI rankings button
button = driver.find_element_by_xpath('//ul[@class="inline-list"]/li[2]/a')
# Clicking on ODI Button
button.click()
#time.sleep(5)
# Locating Full table button for ODI batsman
button = driver.find_element_by_xpath('//div[@class="col-4 col-12-desk uniform-grid__section touch-scroll-list__element"][1]/div/div[2]/a')
# Clicking on full table button
button.click()
#time.sleep(5)
```

```python
# Initialising Batsmen dictionary
batsmen = {}
batsmen["Batsman"] = []
batsmen["Team"] = []
batsmen["Ranking"] = []
```

```python
# Extracting Top 10 Batsmen
for i in range(10):
    try:
        rank = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[1]'.format(i+1))
        batsmen["Ranking"].append(rank.text)
    except NoSuchElementException:
        batsmen["Ranking"].append('-')
            
    try:
        bat = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[2]'.format(i+1))
        batsmen["Batsman"].append(bat.text)
    except NoSuchElementException:
        batsmen["Batsman"].append('-')
    
    try:
        bat = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[3]'.format(i+1))
        batsmen["Team"].append(bat.text)
    except NoSuchElementException:
        batsmen["Team"].append('-')
```

```python
# Locating player Rankings url from website footer Menu
rank_url = driver.find_element_by_xpath('//ul[@class="global-footer__list"][3]/li[3]/a').get_attribute('href')
# Opening Player Rankings
driver.get(rank_url)
time.sleep(5)
# Locating ODI rankings button
button = driver.find_element_by_xpath('//ul[@class="inline-list"]/li[2]/a')
# Clicking on ODI Button
button.click()
# Locating Full table button for ODI bowler
button = driver.find_element_by_xpath('//div[@class="col-4 col-12-desk uniform-grid__section touch-scroll-list__element"][2]/div/div[2]/a')
# Clicking on full table button
button.click()
```

```python
# Initialising Bowler dictionary
bowler = {}
bowler["Bowler"] = []
bowler["Team"] = []
bowler["Ranking"] = []
```

```python
# Extracting Top 10 Bowler
for i in range(10):
    try:
        rank = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[1]'.format(i+1))
        bowler["Ranking"].append(rank.text)
    except NoSuchElementException:
        bowler["Ranking"].append('-')
            
    try:
        bowl = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[2]'.format(i+1))
        bowler["Bowler"].append(bowl.text)
    except NoSuchElementException:
        bowler["Bowler"].append('-')
    
    try:
        bowl = driver.find_element_by_xpath('//table[@class="table rankings-table"]/tbody/tr[{}]/td[3]'.format(i+1))
        bowler["Team"].append(bowl.text)
    except NoSuchElementException:
        bowler["Team"].append('-')
```

```python
teams = pd.DataFrame.from_dict(teams)
batsmen = pd.DataFrame.from_dict(batsmen)
bowler = pd.DataFrame.from_dict(bowler)
```
