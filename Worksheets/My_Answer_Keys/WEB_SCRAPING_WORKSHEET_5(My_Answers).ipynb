1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
# !pip install bs4
# !pip install urllib

from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html, "html.parser")
titles = bs.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
print('ALL HEADER TAGS ::', *titles, sep='\n')
```

2. Write a python program to display IMDB’s Top rated 100 movies’ data(i.e. Name, IMDB rating, duration, director, star cast - 10 main characters)

```python
# !pip install IMDbPY

import pandas as pd
import imdb
ia = imdb.IMDb()

cast_list, title, rating, duration, director = [], [], [], [], []

search = ia.get_top250_movies()
for i in range(100): 
    title.append(search[i]['title']) 
    rating.append(ia.get_movie(search[i].movieID).data['rating'])
    duration.append(ia.get_movie(search[i].movieID).data['runtimes'])
    director.append([d['name'] for d in ia.get_movie(search[i].movieID)['director']])
    cast = ia.get_movie(search[i].movieID).get('cast')[:10]
    cast_list.append([actor['name'] for actor in cast])
    
pd.DataFrame({'Title':title, 'Rating':rating, 'Duration':duration, 'Director':director, 'Cast': cast_list})
```

4. Write a python program to scrap book name, author name, genre and book review of 5 books from ‘www.bookpage.com’

```python
# !pip install bs4
# !pip install urllib

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

html = urlopen('https://bookpage.com/features')
bs = BeautifulSoup(html, "html.parser")

name_list = bs.findAll('h4', attrs={'class':'italic'})
names = [i.text.replace('\n','') for i in name_list[:5]]

author_list = bs.findAll('p', attrs={'class':'sans bold'})
authors = [i.text.replace('\n','').split('by ')[1] for i in author_list[:5]]

genre_list = bs.findAll('p', attrs={'class':'genre-links hidden-phone'})
genre = [i.text.replace('\n','') for i in genre_list[:5]]

review_list = bs.findAll('p', attrs={'class':'excerpt'})
reviews = [i.text.replace('\n','') for i in review_list[:5]]

pd.DataFrame({'Book Name':names, 'Author':authors, 'Genre':genre, 'Review':reviews})
```

5. Write a python program to scrap cricket rankings from ‘www.icc-cricket.com’. You have to scrap:<br>
i) Top 10 ODI teams<br>
ii) Top 10 ODI Batsmen<br>
iii) Top 10 ODI bowlers

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

# i) Top 10 ODI teams
html = urlopen('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
bs = BeautifulSoup(html, "html.parser")
team_list = bs.findAll('span', attrs={'class':'u-hide-phablet'})
teams = [i.text for i in team_list[:10]]

# ii) Top 10 ODI Batsmen
html = urlopen('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
bs = BeautifulSoup(html, "html.parser")
batsmen = []
batsmen_list = bs.findAll('div', attrs={'class':'rankings-block__banner--name'})
batsmen.append(batsmen_list[0].text)
batsmen_list = bs.findAll('td', attrs={'class':'table-body__cell name'})
batsmen = batsmen + [i.text.replace('\n','') for i in batsmen_list[:9]]


# iii) Top 10 ODI bowlers
html = urlopen('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
bs = BeautifulSoup(html, "html.parser")
bowler = []
bowler_list = bs.findAll('div', attrs={'class':'rankings-block__banner--name'})
bowler.append(bowler_list[1].text)
bowler_list = bs.findAll('td', attrs={'class':'table-body__cell name'})
bowler = bowler + [i.text.replace('\n','') for i in bowler_list[9:18]]

df = pd.DataFrame({'Top 10 ODI teams':teams, 'Top 10 ODI Batsmen':batsmen, 'Top 10 ODI bowlers':bowler}, index=range(1,11))
df.index.name = 'RANK'
df
```
