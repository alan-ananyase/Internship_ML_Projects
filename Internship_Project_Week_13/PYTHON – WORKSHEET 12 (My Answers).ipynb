```python
df = pd.read_csv('data_PYTHON_WORKSHEET_12.csv', parse_dates=['Date'])
data = df.copy() # Backup df
df.head(5)
```

```python
# Determine the no. of rows and no. of columns in the dataset.
print('Number of rows ::', df.shape[0])
print('Number of columns ::', df.shape[1])
```

```python
# Display the datatypes of each column.
df.info()
```

```python
# Determine the row indexes of rows having all null values (i.e. empty rows).
for i in df:
    print('>>> Row indexes of null values in', i)
    print(df[df[i].isnull()].index.tolist(), '\n')
```

```python
# Determine the total number of empty rows and their percentage in whole dataset and remove them.
data = df.copy() # Backup df

a = round(data.dropna(axis=0).shape[0] / len(data)*100,2)
print('Number of Empty Rows ::', len(data)-data.dropna(axis=0).shape[0])
print('Percentage of Empty Rows ::', (100-a), '%')

data.dropna(axis=0, inplace=True)
print('>>>> All rows with null values REMOVED')
```

```python
# Determine the column names having missing values.
print([i for i in df if df[i].isna().sum()>0])
```

```python
# Determine the missing values and their percentage in each column.
print('>> Number & Percentage of Empty Rows in the dataset <<')
for i in df:
    print(i, '>>', df[i].isna().sum(), '>>', round((df[i].isna().sum())/len(df)*100,2), '%')
```

```python
# Determine the row indexes having missing values.
print(pd.isnull(df).any(1).nonzero()[0])
```

```python
# Determine the missing values and their percentage in each row.
a = df.shape[0]
for i in range(a):
    print("Total NaN in row", i, "::", df.iloc[i].isnull().sum(), '>>', round((df.iloc[i].isnull().sum()/a)*100,5), '%')
```

```python
# Determine the column names having more than 30% missing values and remove them from the dataset.
data = df.copy() # Backup df

a = 0.3*data.shape[0]
list_30 = [i for i in data if data[i].isna().sum()>a]
print('Columns with more than 30% missing values >>', list_30)
data.drop(columns=list_30, inplace=True)
print('\nColumns REMOVED')
data.head(2)
```

```python
# Determine the number and percentage of rows having more than 5 missing values. If the percentage of such rows is less than 20% then remove all these rows.
data = df.copy()

a = [i for i in range(data.shape[0]) if data.iloc[i].isna().sum()>5]
percent_5missing = round(len(a)/data.shape[0]*100,2)

print('Rows with >5 missing values ::', len(a))
print('Percent of rows with >5 missing values ::', percent_5missing, '%\n')

if percent_5missing<20:
    print('Number of rows before removal', data.shape[0])
    data.drop(index=a, inplace=True)
    data.reset_index(inplace=True)
    print('Number of rows after removal', data.shape[0])
else:
    print('% of rows greater than 20%. **NO ROWS REMOVED**')
```

```python
# Display the description of data and percentage of missing values still left column wise.
for i in df:
    print('###', i, '###')
    print(df[i].describe())
    print('Percentage of missing values ::', round((df[i].isna().sum())/len(df)*100,2), '%\n')
```

```python
# Considering the 'Price' column as target variable which is to be predicted, how will you treat the missing values? Give a brief explanation and treat the missing values in 'Price' column.

# Answer: Since 'Price' is my target variable, I wouldn't want to keep them in my dataset as we don't have values that the Linear model can correctly predict. I would remove the missing rows in my target variable.
data = df.copy() #Backup df
data = data[~(data['Price'].isna())]
data.head(3)
```

```python
# Now check the percentage of missing values in each column again, if missing values exist, treat it with the most suitable method.

# Since 'BuildingArea', 'YearBuilt' & 'CouncilArea' has many missing values, we will first remove those rows and treat other missing values.
data = data[~(data['BuildingArea'].isna() | data['YearBuilt'].isna() | data['CouncilArea'].isna())]
data.reset_index(inplace=True) # Reset index
data.drop(columns=['index'], axis=1, inplace=True)

# Since 'Lattitude' & 'Longtitude' will be the same for the same 'CouncilArea'. We can replace it as per other 'CouncilArea'.
lat_index = data[data['Lattitude'].isna()].index # Find missing values index for 'Lattitude'

for k in lat_index:
    a = data['CouncilArea'][k]
    for i in range(data.shape[0]):
        if data['CouncilArea'][i]==a:
            data['Lattitude'][k]=data['Lattitude'][i]
            data['Longtitude'][k]=data['Longtitude'][i]
data.isna().sum()
```

```python
# Now since you have completed your data cleaning, determine the number and percentage of columns removed.
print('Total number of columns before cleaning::', len(df.columns))
print('Total number of columns after cleaning::', len(data.columns))
a = len(df.columns)-len(data.columns)
print('Total Number of Columns Removed >>', a)
print('Percentage of Columns Removed >>', a*100, '%')
```

```python
# Determine the number and percentage of rows removed in whole process. Also comment briefly that according to you the amount of data lost is justified or not.
print('Number of Rows before cleaning >>', df.shape[0])
print('Number of Rows after cleaning >>', data.shape[0])
print('Total Number of Rows REMOVED >>', df.shape[0]-data.shape[0])
print('Percentage of Rows REMOVED >>', round(data.shape[0]/df.shape[0]*100,2), '%')

# Answer: It is not necessary to remove so much data from the dataset. We would be able to manage missing values also by replacing them with mean and median. This would save more data and helps in training the model perfectly.
```
